## ROC

```python
# Step 2: Sort scores and corresponding labels
sorted_indices = np.argsort(y_pred.flatten())
sorted_scores = y_pred[sorted_indices].flatten()
sorted_labels = y_true[sorted_indices].flatten()

# Step 3: Calculate TPR and FPR at each threshold
tpr = []
fpr = []
P = np.sum(y_true == 1)  # Total positives
N = np.sum(y_true == 0)  # Total negatives

for score in sorted_scores:
    # Calculate the number of true positives and false positives at this threshold
    TP_at_threshold = np.sum((sorted_labels == 1) & (sorted_scores >= score))
    FP_at_threshold = np.sum((sorted_labels == 0) & (sorted_scores >= score))

    # Calculate TPR and FPR at this threshold
    tpr_at_threshold = TP_at_threshold / P if P != 0 else 0
    fpr_at_threshold = FP_at_threshold / N if N != 0 else 0

    tpr.append(tpr_at_threshold)
    fpr.append(fpr_at_threshold)

tpr = np.array(tpr + [0])
fpr = np.array(fpr + [0])

-np.trapz(tpr, fpr)
```

```python
px.line(x=fpr, y=tpr).show()
```

```python
# Step 3: Calculate Precision and Recall at each threshold
precision = []
recall = []
P = np.sum(y_true == 1)  # Total positives

TP = 0
FP = 0

for score in sorted_scores:
    # Calculate the number of true positives and false positives at this threshold
    TP_at_threshold = np.sum((sorted_labels == 1) & (sorted_scores >= score))
    FP_at_threshold = np.sum((sorted_labels == 0) & (sorted_scores >= score))

    # Calculate Recall at this threshold
    recall_at_threshold = TP_at_threshold / P if P != 0 else 0

    # Calculate Precision at this threshold
    precision_at_threshold = TP_at_threshold / (TP_at_threshold + FP_at_threshold) if (TP_at_threshold + FP_at_threshold) != 0 else 0

    recall.append(recall_at_threshold)
    precision.append(precision_at_threshold)

# Step 4: Convert lists to numpy arrays
precision = np.array(precision)
recall = np.array(recall)
-np.trapz(tpr, fpr)
```

```python
px.line(x=precision, y=recall).show()
```
